```python
df1 = df.copy()
for i in range(len(df1['class'])):
    if(df1['class'][i]=='Iris-setosa'):
        df1['class'][i] = 0
    elif(df1['class'][i]=='Iris-versicolor'):
        df1['class'][i] = 1
    else:
        df1['class'][i] = 2
df1
```

```python
corr_matrix = df1.corr()
corr_matrix['petal-width'].sort_values(ascending=False)
```

```python
scatter_matrix(df1[['petal-width','sepal-width','petal-length','sepal-length']],figsize=(8,8))
```

```python
sss = StratifiedShuffleSplit(n_splits = 5, test_size = 0.3, random_state=42)
for train_index,test_index in sss.split(df,df['class']):
    train_set = df.loc[train_index]
    test_set = df.loc[test_index]
```

```python
data_train = train_set.drop(['class'],axis=1)
labels_train = train_set['class']
data_test = test_set.drop(['class'],axis=1)
labels_test = test_set['class']
```

```python
kfold = KFold(n_splits = 5,random_state=42,shuffle=True)
for i in range(3,15):
    model = KNeighborsClassifier(n_neighbors=i,weights='distance')
    result = cross_val_score(model,data_train,labels_train,cv=kfold,scoring='accuracy')
    print(i,":",result.mean())
```

```python
model = KNeighborsClassifier(n_neighbors=10,weights='distance')
model.fit(data_train,labels_train)
prediction = model.predict(data_test)
```

```python
classification_report(labels_test,prediction)
```

```python
confusion_matrix(labels_test,prediction)
```

```python
kfold = KFold(n_splits = 5,random_state=42,shuffle=True)
for i in range(3,15):
    model = KNeighborsClassifier(n_neighbors=i,weights='uniform')
    result = cross_val_score(model,data_train,labels_train,cv=kfold,scoring='accuracy')
    print(i,":",result.mean())
```
